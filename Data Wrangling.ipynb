### 2017, Top 5 countries with the most cell phones total?

```python
##1st method
top = joined_data[joined_data['year'] == 2017]
top_five = top[['Country', 'cell_phones_total']].groupby('Country').first()
sort = top_five.sort_values('cell_phones_total', ascending = False)
sort.head(5)
```

```python
# 2nd method
top= joined_data[joined_data['year']==2017]
df=pd.DataFrame()
df['Country']=top['Country']
df['cell_phones_total']=top['cell_phones_total']

df.sort_values("cell_phones_total", inplace = True, ascending=False) 
df.head(5)
```
